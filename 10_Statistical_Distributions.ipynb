## Statistical Distributions

In data analysis, we often deal with data distributions, which help us understand how data points are spread out. One of the most commonly used distributions is the Gaussian distribution, also known as the Normal distribution.

- The Gaussian distribution is a mathematical function that we can use to figure out the probability of any specific data point occurring within our dataset.
- It essentially describes how data points are grouped or how dense they are, which we call the probability density function.
- We can also use it to calculate the likelihood of a data point having a value equal to or less than a given value.
- This overall summary of these data relationships is referred to as a cumulative density function.

In this tutorial, we'll explore the Gaussian distribution and other related distributions, like the Student’s t and Chi-Squared distributions, which are connected to the Gaussian distribution.

By the end of this tutorial, you'll gain the following knowledge:

1. An easy-to-understand introduction to standard distributions, helping you grasp how data relationships work.
2. How to calculate and create plots for probability and density functions related to the Gaussian distribution.
3. An overview of the Student’s t and Chi-Squared distributions, which are associated with the Gaussian distribution.

## Distributions

- In practical terms, a distribution can be thought of as a function that defines the connection between observations within a sample space.
  - For instance, consider human ages, where each age serves as an observation within the domain, and the sample space extends from ages 0 to 125.
  - This distribution is a mathematical function that outlines how observations are distributed across various ages.

- A distribution essentially represents a collection of data or scores associated with a variable.
  - Typically, these scores are organized from the smallest to the largest and can be visually depicted in graphs.

- As explained in "Statistics in Plain English, Third Edition" (2010, Page 6), many datasets adhere to well-known mathematical functions, such as the Gaussian distribution.
  - By adjusting the parameters of these functions, such as the mean and standard deviation in the case of the Gaussian, the function can be tailored to fit the data.
  - Once a distribution function is established, it becomes a convenient tool for describing and computing related quantities, including
    - the likelihood of specific observations and
    - plotting the relationship among observations within the domain.

## Density Functions

Distributions are often described in terms of their density functions, which provide insight into how the proportion of data or the likelihood of observations changes across the distribution's range. There are two main types of density functions:

- **Probability Density Function (PDF):** This function helps calculate the probability of observing a specific value.
  - It can also be employed to summarize the likelihood of observations across the entire sample space of the distribution.
  - When visualized, PDF plots typically exhibit the recognizable shape of a distribution, such as the bell-curve for the Gaussian distribution.
  - Distributions are frequently characterized in terms of their probability density functions and their associated parameters.

- **Cumulative Density Function (CDF):** In contrast to the PDF, the CDF calculates the cumulative likelihood for an observation and all preceding observations within the sample space.
  - It allows for a quick understanding of how much of the distribution falls before or after a given value.
  - CDFs are often graphically represented as a curve ranging from 0 to 1, showcasing the distribution's behavior.

- Both PDFs and CDFs are continuous functions. For discrete distributions, the equivalent of a PDF is termed a Probability Mass Function (PMF).

- In the following sections, we will delve into the Gaussian distribution and two related distributions that are commonly encountered in statistical analyses. We will explore each of these distributions in terms of their parameters, probability functions, and cumulative density functions.

# Gaussian Distribution

- The Gaussian distribution, named after Carl Friedrich Gauss, holds a central position in statistics due to its widespread applicability.
  - Remarkably, data from various fields often conforms to a Gaussian distribution, making it so prevalent that it's commonly referred to as the "normal distribution."

- Key parameters used to describe a Gaussian distribution are:

  - **Mean (μ):** This represents the expected value of the distribution.

  - **Variance (σ²):** The variance, denoted with the Greek lowercase letter sigma (σ²), quantifies the spread of observations around the mean.

- To facilitate calculations, it's common to use the standardized form of the variance called the **standard deviation (σ):** This parameter characterizes the normalized spread of observations from the mean.

- In practice, we can work with the Gaussian distribution using the `norm` module in SciPy.
  - The `norm.pdf()` function is employed to create a Gaussian Probability Density Function (PDF) by specifying the sample space, mean, and standard deviation.
  - For example, the code snippet below generates a Gaussian PDF with a sample space from -5 to 5, a mean of 0, and a standard deviation of 1.
  - Such a Gaussian with a mean of 0 and a standard deviation of 1 is termed the "Standard Gaussian."

- This code generates a line plot, with the x-axis representing the sample space and the y-axis indicating the likelihood of each value.
  - The bell-shaped curve is characteristic of the Gaussian distribution, with the peak signifying the most probable value, known as the expected value or the mean.

---


```python
# plot the Gaussian PDF
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
mean = 0.0
stdev = 1.0

# calculate the PDF
pdf = norm.pdf(sample_space, mean, stdev)

# plot
pyplot.plot(sample_space, pdf)
pyplot.show()
```

- Similarly, we can create a Gaussian Cumulative Density Function (CDF) using the `norm.cdf()` function, as shown in the following code snippet:

  - Running this code generates a plot showcasing an S-shaped curve, with the x-axis representing the sample space and the y-axis representing cumulative probabilities.
  
  - Notably, a value of 2 covers nearly 100% of the observations, with only a minimal tail of the distribution extending beyond that point.

  - Additionally, the mean value of zero signifies that 50% of the observations lie before and after this point.

```python
# plot the Gaussian CDF
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# define the distribution parameters
sample_space = arange(-5, 5, 0.001)

# calculate the CDF
cdf = norm.cdf(sample_space)

# plot
pyplot.plot(sample_space, cdf)
pyplot.show()
```

# Student’s t-Distribution

- The Student's t-distribution, often referred to as the t-distribution, is named after its pseudonymous creator, "Student" (William Sealy Gosset).

- This distribution comes into play when estimating the mean of a normal distribution from samples of varying sizes.
  - It proves to be a valuable tool when dealing with uncertainty or error related to estimating population statistics, especially when sample size plays a crucial role.

- While you may not directly utilize the t-distribution in your work, you might need to estimate values from this distribution for other statistical methods, such as significance tests.

- The t-distribution is characterized by a single parameter:

  - **Number of Degrees of Freedom (ν):** Denoted using the lowercase Greek letter nu (ν), it represents the number of degrees of freedom.

- Understanding the appropriate number of degrees of freedom is essential for using the t-distribution.
  - The degrees of freedom depict the amount of information used to describe a population quantity.
  - For instance, when calculating the mean, there are n degrees of freedom because all n observations in the sample are employed to estimate the population mean.
  - When a statistical quantity depends on another in its calculation, one must subtract 1 from the degrees of freedom, as is the case with the mean in the computation of sample variance.

- Student’s t-distribution observations are derived from Gaussian distribution observations to describe the interval for the population mean in the normal distribution. These observations are calculated as follows:

  $ data = \frac{x - \text{mean}(x)}{\sqrt{S/n}} $

  - Where x represents observations from the Gaussian distribution, mean denotes the average observation of x,
  - S represents the standard deviation, and n is the total number of observations.
  - The resulting observations constitute the t-observation with (n - 1) degrees of freedom.
  - In practice, if you need a value from a t-distribution when calculating a statistic, the number of degrees of freedom is likely to be (n - 1), where n is the size of your sample drawn from a Gaussian distribution.

- The choice of a specific distribution for a given problem hinges on the size of your sample. - Statistics in Plain English, Third Edition (Page 93, 2010)

  ---

- You can work with the t-distribution in Python using the SciPy library's `stats.t` module.
  - The `t.pdf()` function allows you to generate a Student’s t-distribution with specified degrees of freedom.
  - In the example below, we create a t-distribution with a sample space ranging from -5 to 5 and (10,000 - 1) degrees of freedom.

- Running this code generates a plot showing the t-distribution PDF, which resembles the familiar bell shape of the normal distribution.
  - A noticeable difference is the fatter tails in the distribution, indicating a higher likelihood of observations in the tails compared to the Gaussian distribution.


---


```python
# plot the t-distribution PDF
from numpy import arange
from matplotlib import pyplot
from scipy.stats import t

# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
dof = len(sample_space) - 1

# calculate the PDF
pdf = t.pdf(sample_space, dof)

# plot
pyplot.plot(sample_space, pdf)
pyplot.show()
```

- You can also create the cumulative density function (CDF) for the t-distribution using the `t.cdf()` function. The code below generates a CDF over the same range as before:

  - Running this code results in an S-shaped curve, similar to that of the Gaussian distribution, with slightly smoother transitions from zero-probability to one-probability due to the fatter tails of the t-distribution.

```python
# plot the t-distribution CDF
from numpy import arange
from matplotlib import pyplot
from scipy.stats import t

# define the distribution parameters
sample_space = arange(-5, 5, 0.001)
dof = len(sample_space) - 1

# calculate the CDF
cdf = t.cdf(sample_space, dof)

# plot
pyplot.plot(sample_space, cdf)
pyplot.show()
```

# Chi-Squared Distribution

- The Chi-Squared distribution is represented by the lowercase Greek letter chi (χ) raised to the power of 2 (χ²).
  - For ease of reference, we typically write it as Chi-Squared.
  - Much like the Student’s t-distribution, the Chi-Squared distribution finds application in statistical methods when dealing with data drawn from a Gaussian distribution to quantify uncertainty.
  - For instance, the Chi-Squared distribution plays a pivotal role in Chi-Squared statistical tests for independence and is also involved in the derivation of the Student’s t-distribution.

- The Chi-Squared distribution is defined by a single parameter:

  - **Degrees of Freedom (k):** Denoted as 'k,' this parameter characterizes the number of degrees of freedom.

- An observation in a Chi-Squared distribution is calculated as the sum of the squares of 'k' observations drawn from a Gaussian distribution:

  $ chi = \sum_{i=1}^{k} x_i^2 $

  - Here, χ represents an observation with a Chi-Squared distribution,
  - 'x' denotes observations drawn from a Gaussian distribution, and
  - 'k' signifies the number of 'x' observations, which is also equivalent to the number of degrees of freedom for the Chi-Squared distribution.
  - Similar to the Student’s t-distribution, data is not inherently fit for a Chi-Squared distribution; instead, observations are drawn from this distribution in the calculation of statistical methods for a sample of Gaussian data.

  ---

- You can leverage the `stats.chi2` module in SciPy to perform Chi-Squared distribution calculations.
  - The `chi2.pdf()` function allows you to compute the Chi-Squared distribution for a sample space ranging from 0 to 50 with 20 degrees of freedom.

  - Note that squared values must be positive, which is why a positive sample space is required.

  - Running this code generates the Chi-Squared PDF as a line plot.
    - With 20 degrees of freedom, the expected value of the distribution falls just short of the value 20 on the sample space.
    - This is intuitive, considering that most of the density in the Gaussian distribution lies between -1 and 1.
    - Consequently, the sum of squared random observations from the standard Gaussian would sum to just under the number of degrees of freedom, in this case, 20 (approximately 20 × 1).
    - Despite the bell-like shape, the Chi-Squared distribution is not symmetric.



---

```python
# plot the Chi-Squared PDF
from numpy import arange
from matplotlib import pyplot
from scipy.stats import chi2

# define the distribution parameters
sample_space = arange(0, 50, 0.01)
dof = 20

# calculate the PDF
pdf = chi2.pdf(sample_space, dof)

# plot
pyplot.plot(sample_space, pdf)
pyplot.show()
```

- You can also calculate the cumulative density function (CDF) for the Chi-Squared distribution using the `chi2.cdf()` function. The code below demonstrates creating a CDF over the same sample space:

  - Executing this code results in a plot of the cumulative density function for the Chi-Squared distribution, showcasing the likelihood of the Chi-Squared value around 20 with the fat tail extending to the right of the distribution, beyond the end of the plot.

```python
# plot the Chi-Squared CDF
from numpy import arange
from matplotlib import pyplot
from scipy.stats import chi2

# define the distribution parameters
sample_space = arange(0, 50, 0.01)
dof = 20

# calculate the CDF
cdf = chi2.cdf(sample_space, dof)

# plot
pyplot.plot(sample_space, cdf)
pyplot.show()
```
