# Introduction to Matplotlib

Python offers numerous libraries for creating visually appealing plots and graphics. If you're looking for a quick and straightforward way to create plots, especially for personal use, I recommend using the Matplotlib library. Matplotlib serves as the foundation for many other plotting libraries and is integrated into higher-level libraries like Pandas.

- Matplotlib operates within a context, where you can create one or more plots before displaying or saving them. You can access this context through functions provided by the `pyplot` module. Typically, this context is imported as follows:

  ```python
  # Import the Matplotlib context
  from matplotlib import pyplot
  ```

  - While many choose to import the context as `plt`, we'll stick to the standard Python import convention.

- To create charts and plots, you simply make and call functions within this context. For example:

  ```python
  # Create a plot
  pyplot.plot(...)
  ```

- You can configure elements such as axes, labels, legends, and more using separate function calls within the context. To view the drawings, you can use the `show()` function:

  ```python
  # Display the plot
  pyplot.show()
  ```

- Alternatively, you can save your plots as image files, such as PNG, using the `savefig()` function:

  ```python
  # Save the plot to a file
  pyplot.savefig('my_image.png')
  ```

# Line Plot

- A line plot is a useful way to visualize data that is collected at regular intervals.
- In a line plot, the x-axis typically represents the regular intervals, such as time, while the y-axis displays the observations ordered by the x-axis and connected with a line.
- To create a line plot, you can use the `plot()` function by providing the x-axis data for the regular interval and the y-axis data for the observations.

  ```python
  # Create a line plot
  pyplot.plot(x, y)
  ```

- Line plots are particularly handy for presenting time series data and any other sequence data where there is a natural order between observations.

- Let's consider an example where we create a sequence of 100 floating-point values for the x-axis and use a sine wave function to generate observations for the y-axis. The results are then plotted as a line plot below.

  - When you run this example, it will produce a line plot that displays the familiar sine wave pattern on the y-axis across the x-axis with a consistent interval between observations.

````python
# Example of a line plot
from numpy import sin
from matplotlib import pyplot

# Consistent interval for the x-axis
x = [x * 0.1 for x in range(100)]

# Function of x for the y-axis
y = sin(x)

# Create a line plot
pyplot.plot(x, y)

# Show the line plot
pyplot.show()
````

# Bar Chart

- A bar chart is commonly used to represent relative quantities for multiple categories.
  - In a bar chart, the x-axis represents these categories, evenly spaced, while the y-axis represents the quantity associated with each category.
  - Each quantity is depicted as a bar extending from the baseline to the appropriate level on the y-axis.
- To create a bar chart, you can use the `bar()` function, passing the category names for the x-axis and the quantities for the y-axis.

  ```python
  # Create a bar chart
  pyplot.bar(x, y)
  ```

- Bar charts are effective for comparing multiple values or estimates across different categories.

- In the following example, we have a dataset with three categories, each identified by a string label. A random integer value is assigned to each category:

  - Running this example will generate a bar chart displaying the category labels on the x-axis and the corresponding quantities on the y-axis.

````python
# Example of a bar chart
from random import seed
from random import randint
from matplotlib import pyplot

# Seed the random number generator
seed(1)

# Category names
x = ['red', 'green', 'blue']

# Quantities for each category
y = [randint(0, 100), randint(0, 100), randint(0, 100)]

# Create the bar chart
pyplot.bar(x, y)

# Show the bar chart
pyplot.show()
````

# Histogram Plot

A histogram plot is commonly used to summarize the distribution of a data sample.
- In a histogram, the x-axis represents discrete bins or intervals for observations.
  - For instance, observations with values between 1 and 10 may be divided into five bins. The values [1,2] would belong to the first bin, [3,4] to the second bin, and so on.

- The y-axis represents the frequency or count of observations in the dataset that fall into each bin.

- Essentially, a data sample is transformed into a bar chart where each category on the x-axis represents an interval of observation values.

- Histograms are valuable tools for providing density estimates, offering a visual representation of data distribution. As the data is counted within consecutive intervals (bins), it helps to understand the overall shape and pattern of the data.

    > "Histograms are density estimates. A density estimate gives a good impression of the distribution of the data."
    > — Applied Multivariate Statistical Analysis, 2015

  ---

- To create a histogram plot, you can use the `hist()` function by passing a list or array that represents the data sample.

  ```python
  # Create a histogram plot
  pyplot.hist(x)
  ```

- Histograms are particularly useful for summarizing data distributions. In the following example, a dataset of 1,000 random numbers drawn from a standard Gaussian distribution is created and then plotted as a histogram:

  - Running this example reveals a bell-shaped curve representing the Gaussian distribution. The function automatically determines the number of bins to display the data, splitting values into groups based on integer values.


  - You can further fine-tune the histogram by specifying the number of bins using the `bins` argument, as demonstrated below:

  ```python
  # Create a histogram plot with a specific number of bins
  pyplot.hist(x, bins=100)
  ```




````python
# Example of a histogram plot
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot

# Seed the random number generator
seed(1)

# Generate random numbers from a Gaussian distribution
x = randn(1000)

# Create a histogram plot
pyplot.hist(x)

# Show the histogram plot
pyplot.show()
````

# Box and Whisker Plot

A box and whisker plot, commonly referred to as a boxplot, is a visualization tool used to summarize the distribution of a data sample.
- The x-axis is used to represent the data sample, and you can draw multiple boxplots side by side on the x-axis if needed. The y-axis represents observation values.

- In a boxplot, a box is drawn to summarize the middle 50% of the dataset, starting at the observation at the 25th percentile and ending at the 75th percentile.
- The median, which represents the 50th percentile, is drawn as a line within the box.
- The interquartile range (IQR), calculated as 1.5 times the difference between the 75th and 25th percentiles, is used to determine the length of the box.

---

- Lines called whiskers are drawn from both ends of the box, with a length of the IQR, demonstrating the expected range of sensible values in the distribution.
  - Observations outside the whiskers might be considered outliers and are depicted with small circles.

  > "The boxplot is a graphical technique that displays the distribution of variables. It helps us see the location, skewness, spread, tile length, and outlying points. [...] The boxplot is a graphical representation of the Five Number Summary."
  > — Applied Multivariate Statistical Analysis, 2015

  ---

- You can create boxplots using the `boxplot()` function by passing in the data sample as an array or list.

  ```python
  # Create a box and whisker plot
  pyplot.boxplot(x)
  ```

- Boxplots are valuable for summarizing data distribution and offer an alternative to histograms. They provide a quick overview of the range of common and reasonable values both within the box and the whisker.
- This method is often used when the data has an unknown or unusual distribution, such as non-Gaussian.

---

- In the example below, we create three box and whisker plots in a single chart.
  - Each plot summarizes a data sample drawn from a slightly different Gaussian distribution. We generate each data sample as an array and add all three data sample arrays to a list before passing it to the plotting function.

  - Running this example generates a chart displaying the three box and whisker plots. Notice that the same y-axis scale is used for each plot, making the first plot appear compressed and the last plot stretched out.
  - In each plot, you can see the black box representing the middle 50% of the data, the orange line for the median, whisker lines summarizing the range of sensible data, and dots for possible outliers.


````python
# Example of a box and whisker plot
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot

# Seed the random number generator
seed(1)

# Generate random numbers drawn from Gaussian distributions
x = [randn(1000), 5 * randn(1000), 10 * randn(1000)]

# Create a box and whisker plot
pyplot.boxplot(x)

# Show the box and whisker plot
pyplot.show()
````

# Scatter Plot

A scatter plot, or scatterplot, is typically used to visualize the relationship between two paired data samples.
  - Paired data samples mean that two measurements were recorded for a given observation, such as a person's weight and height.
  - In a scatter plot, the x-axis represents the observation values for the first sample, and the y-axis represents the observation values for the second sample. Each point on the plot represents a single observation.

  > "Scatterplots are bivariate or trivariate plots of variables against each other. They help us understand relationships among the variables of a dataset. A downward-sloping scatter indicates that as we increase the variable on the horizontal axis, the variable on the vertical axis decreases."
  > — Applied Multivariate Statistical Analysis, 2015

  ---

- To create a scatter plot, you can use the `scatter()` function by passing the two data sample arrays.

  ```python
  # Create a scatter plot
  pyplot.scatter(x, y)
  ```

- Scatter plots are valuable for demonstrating the association or correlation between two variables.
  - This correlation can be quantified, for example, by adding a line of best fit to the chart, making the relationship clearer.
  - In some cases, a dataset may contain more than two measures (variables or columns) for a given observation.
  - A scatter plot matrix is a collection of scatter plots, each representing a pair of variables in a dataset with more than two variables.

 ---

- In the following example, two data samples are created that are related.
  - The first is a sample of random numbers drawn from a standard Gaussian distribution.

  - The second sample depends on the first by adding a second random Gaussian value to the value of the first measure.

  - Running this example generates a scatter plot, revealing a positive relationship between the two variables.


````python
# Example of a scatter plot
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot

# Seed the random number generator
seed(1)

# First variable
x = 20 * randn(1000) + 100

# Second variable
y = x + (10 * randn(1000) + 50)

# Create a scatter plot
pyplot.scatter(x, y)

# Show the scatter plot
pyplot.show()
````
